```python
# visualize
def use_svg_display():  #@save
    """Use the svg format to display a plot in Jupyter."""
    backend_inline.set_matplotlib_formats('svg')

def set_figsize(figsize=(3.5, 2.5)):  #@save
    """Set the figure size for matplotlib."""
    use_svg_display()
    d2l.plt.rcParams['figure.figsize'] = figsize

#@save
def set_axes(axes, xlabel, ylabel, xlim, ylim, xscale, yscale, legend):
    """Set the axes for matplotlib."""
    axes.set_xlabel(xlabel), axes.set_ylabel(ylabel)
    axes.set_xscale(xscale), axes.set_yscale(yscale)
    axes.set_xlim(xlim),     axes.set_ylim(ylim)
    if legend:
        axes.legend(legend)
    axes.grid()
```

```python
#@save
def plot(X, Y=None, xlabel=None, ylabel=None, legend=[], xlim=None,
         ylim=None, xscale='linear', yscale='linear',
         fmts=('-', 'm--', 'g-.', 'r:'), figsize=(3.5, 2.5), axes=None):
    """Plot data points."""

    def has_one_axis(X):  # True if X (tensor or list) has 1 axis
        return (hasattr(X, "ndim") and X.ndim == 1 or isinstance(X, list)
                and not hasattr(X[0], "__len__"))

    if has_one_axis(X): X = [X]
    if Y is None:
        X, Y = [[]] * len(X), X
    elif has_one_axis(Y):
        Y = [Y]
    if len(X) != len(Y):
        X = X * len(Y)

    set_figsize(figsize)
    if axes is None:
        axes = d2l.plt.gca()
    axes.cla()
    for x, y, fmt in zip(X, Y, fmts):
        axes.plot(x,y,fmt) if len(x) else axes.plot(y,fmt)
    set_axes(axes, xlabel, ylabel, xlim, ylim, xscale, yscale, legend)
```

```python
# plot f(x) = 3*x^2 - 4*x & tangent line (x=1)
x = np.arange(0, 3, 0.1)
plot(x, [f(x), 2 * x - 3], 'x', 'f(x)', legend=['f(x)', 'Tangent line (x=1)'])
```

# Automatic Differentiation

```python
x = torch.arange(4.0)
x
```

```python
# gradient를 계산할 수 있도록 함
x.requires_grad_(True)
x.grad
```

```python
y = 2 * torch.dot(x, x)
y
```

```python
# backward를 통해 gradient를 계산하고 계산한 값은 grad를 통해 접근
y.backward()
x.grad
```

```python
x.grad == 4 * x
```

```python
# 파이토치는 자동으로 gradient 버퍼를 비우지 않기 때문에 버퍼를 비워주어야 함
x.grad.zero_() # Reset the gradient
y = x.sum()
y.backward()
x.grad
```

```python
# gradient를 계산하기 전 y-dimensional 텐서로 초기화
x.grad.zero_()
y = x * x
y.backward(gradient=torch.ones(len(y)))  # Faster: y.sum().backward()
x.grad
```

```python
# detach()를 통해 해당 변수를 gradient 계산에서 제외
x.grad.zero_()
y = x * x
u = y.detach() # u를 상수로 생각(미분 대상으로 생각 X)
z = u * x

z.sum().backward()
x.grad == u
```

```python
# 단, 여전히 x에 대해서는 gradient 계산 가능
x.grad.zero_()
y.sum().backward()
x.grad == 2 * x
```

```python
def f(a):
    b = a * 2
    while b.norm() < 1000:
        b = b * 2
    
    if b.sum() > 0:
        c = b
    else:
        c = 100 * b
        
    return c
```

```python
a = torch.randn(size=(), requires_grad=True)
d = f(a)
d.backward()
```

```python
# linear 함수이므로 gradient는 단순히 변수에 해당하는 a로 나눈 값
a.grad == d / a
```

# Probability & Statistics

```python
%matplotlib inline
import random
import torch
from torch.distributions.multinomial import Multinomial
from d2l import torch as d2l
```

## Coin Toss 예제

- Multinomial() : 나올 수 있는 출력에 해당하는 확률을 통해 특정 횟수만큼 시뮬레이션
- 각 경우의 수에 대한 frequency를 벡터로 반환

```python
num_tosses = 100
heads = sum([random.random() > 0.5 for _ in range(num_tosses)])
tails = num_tosses - heads
print("heads, tails: ", [heads, tails])
```

```python
fair_probs = torch.tensor([0.5, 0.5])
Multinomial(100, fair_probs).sample() # frequency of each outcome
```

```python
# 확률로 만들기 위해 전체 횟수로 나누어줌
Multinomial(100, fair_probs).sample() / 100
```

```python
# 샘플 크기가 증가하면, 0.5라는 수렴값에 더욱 가까워진다는 것을 확인할 수 있음
counts = Multinomial(10000, fair_probs).sample()
counts / 10000 
```

```python
# 샘플 크기가 n배 증가할수록, error는 (1/sqrt(n))의 비율로 감소
counts = Multinomial(1, fair_probs).sample((10000,))
cum_counts = counts.cumsum(dim=0) # cumulative sum
estimates = cum_counts / cum_counts.sum(dim=1, keepdims=True)
estimates = estimates.numpy()

# visualize
d2l.set_figsize((4.5, 3.5))
d2l.plt.plot(estimates[:, 0], label=("P(coin=heads)"))
d2l.plt.plot(estimates[:, 1], label=("P(coin=tails)"))
d2l.plt.axhline(y=0.5, color='black', linestyle='dashed')
d2l.plt.gca().set_xlabel('Samples')
d2l.plt.gca().set_ylabel('Estimated probability')
d2l.plt.legend()
```
